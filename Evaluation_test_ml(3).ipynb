```python
# 5)Fill the missing values in Married,Gender,Self_Employed with randint, LoanAmount and Dependents with median,Loan_Amount_Term with mean ?
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
rndval = np.random.randint(5,10)
train_df['Gender'].unique()
train_df['Gender'].fillna(rndval,inplace=True)
train_df['Gender'].unique()
train_df['Married'].fillna(rndval,inplace=True)
train_df['Self_Employed'].fillna(rndval,inplace=True)
train_df['Self_Employed'].unique()
mdan = train_df['LoanAmount'].median()
train_df['LoanAmount'].fillna(mdan,inplace=True)
mean =  train_df['Loan_Amount_Term'].mean()
train_df['Loan_Amount_Term'].fillna(mean,inplace=True)
train_df['Loan_Amount_Term'].unique()
```
